```python
#!pip install yfinance
#!pip install pandas

# Importando bibliotecas que precisarei

# biblioteca do Yahoo Finance
import yfinance as yf
import pandas as pd

# Importando o que precisaremos especificamente
# https://www.tutorialspoint.com/matplotlib/matplotlib_pylab_module.htm
import matplotlib.pyplot as plt
```

```python
serie_renner = yf.download('LREN3.SA', start='2019-01-01', end='2023-02-02', rounding=True)
serie_lvmh = yf.download('MC.PA', start='2019-01-01', end='2023-02-02', rounding=True)
serie_nike = yf.download('NKE', start='2019-01-01', end='2023-02-02', rounding=True)
```

```python
plt.figure(figsize=(20,10),facecolor='white')
plt.plot(serie_lvmh['Adj Close'], 'r--')
plt.title('STOCKS', fontsize=30, loc='Center')
plt.xlabel('Data', fontsize=15)
plt.ylabel('Preo', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(serie_renner['Adj Close'])
plt.plot(serie_lvmh['Adj Close'])
plt.plot(serie_nike['Adj Close'])
```

```python
df = pd.DataFrame((serie_renner['Adj Close'],serie_lvmh['Adj Close'],serie_nike['Adj Close'])).T

df.head()
```

```python
df.columns = ['Fech Renner', 'Fech LV', 'Fech Nike']
df.head()
```

```python
print(df.iloc[0])
```

```python
# Ele extrair o primeiro valor da tabela.

df.iloc[0]

plt.figure(figsize=[20,10],facecolor='white')
plt.plot((df['Fech Renner'] / df['Fech Renner'].iloc[0] * 100),color='red')
plt.plot((df['Fech LV'] / df['Fech LV'].iloc[0] * 100),color='brown')
plt.plot((df['Fech Nike'] / df['Fech Nike'].iloc[0] * 100),color='black')
plt.title('Comparao Normalizada a 100',fontsize=15)
plt.legend(['Renner','LVMH','Nike'])
plt.show()
```
