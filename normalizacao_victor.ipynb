```python
serie_renner = yf.download('LREN3.SA', start='2015-01-01', end='2023-02-02', rounding=True)
serie_lvmh = yf.download('MC.PA', start='2015-01-01', end='2023-02-02', rounding=True)
serie_nike = yf.download('NKE', start='2015-01-01', end='2023-02-02', rounding=True)
```

```python
plt.figure(figsize=(20,10),facecolor='white')
plt.plot(serie_lvmh['Adj Close'], 'r--')
plt.title('STOCKS', fontsize=30, loc='Center')
plt.xlabel('Data', fontsize=15)
plt.ylabel('Preo', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(serie_renner['Adj Close'])
plt.plot(serie_lvmh['Adj Close'])
plt.plot(serie_nike['Adj Close'])
```
